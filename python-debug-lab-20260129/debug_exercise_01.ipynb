# Este programa pergunta dados do utilizador.

```python
# Pede os dados do utilizador

nome = input("Qual é o teu nome? ")
idade = int(input("Qual é a tua idade? "))
peso = float(input("Qual é o teu peso em kg? "))
```

```python
# Retorna os dados (código com f-string)
print(f"Olá, {nome}!")
print(f"Tens {idade} anos.")
print(f"Pesas {peso} kg.")
```
