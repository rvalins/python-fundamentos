```python
##Este programa pergunta dados do utilizador.

nome = input("Qual é o teu nome?")
idade = input("Qual é a tua idade?")
peso = input("Qual é o teu peso em kg?")

print("Olá " + nome)  # Converter todos os prints para f-string
print("Tens " + idade + " anos ")
print("Pesas " + peso + " kg")

print(f"Olá {nome}, tens {idade} anos e pesas {peso}kg")
print("Programa terminado")
```
