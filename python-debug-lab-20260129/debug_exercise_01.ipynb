```python
# Este programa pergunta dados do utilizador.

nome = input("Qual é o teu nome? ").strip().title()
idade = int(input("Qual é a tua idade? "))
peso = float(input("Qual é o teu peso em kg? ").replace(",", "."))

print(f"Olá {nome}")
print(f"Tens {idade} anos")
print(f"Pesas {peso} kg")

print("Programa terminado")
```
